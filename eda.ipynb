```python
import json
```

```python
with open("dataset_json/global_metadata.json") as f:
    data = json.load(f)
```

```python
titles = []
h1 = []
h2 = []
h3 = []
```

```python
for d in data:
    titles.append(d["title"])
    for out in d["outline"]:
        if out["level"] == "H1":
            h1.append(out["text"])
        elif out["level"] == "H2":
            h2.append(out["text"])
        else:
            h3.append(out["text"])
```

```python
import matplotlib.pyplot as plt
```

```python
lengths = [len(titles), len(h1), len(h2), len(h3)]
labels = ['Title', 'H1', 'H2', 'H3']
```

```python
plt.bar(labels, lengths, color='skyblue')
plt.xlabel('Heading Levels')
plt.ylabel('Number of Entries')
plt.title('Counts of Each Heading Level')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
import seaborn as sns
import pandas as pd

data = (
    [('Title', len(s.split())) for s in titles] +
    [('H1', len(s.split())) for s in h1] +
    [('H2', len(s.split())) for s in h2] +
    [('H3', len(s.split())) for s in h3]
)

df = pd.DataFrame(data, columns=['Level', 'Length'])

plt.figure(figsize=(12, 6))
sns.kdeplot(data=df, x='Length', hue='Level', fill = True, common_norm=False, alpha=0.5)
plt.title('Length Distribution of Headings (KDE)')
plt.grid(True)
plt.show()
```
